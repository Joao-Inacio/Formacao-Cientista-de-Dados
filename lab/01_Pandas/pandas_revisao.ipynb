```python
# Importando arquivos e base de dados
df_lista = pd.read_csv('../lista_disc.csv')
# display(df_lista)
df_lista.head(10)
```

```python
# .loc pega a linha conforme o índice 
# .loc[linhas, colunas]
df_lista.loc[1].head(19)
```

```python
# Pegando uma linha que corresponde a uma condição
display(df_lista.loc[df_lista['ano_ingresso'] == 2019])
```

```python
# pegando várias linhas e colunas usando o loc
ano2019 = df_lista.loc[df_lista['ano_ingresso'] == 2019, ['periodo_ingresso', 'sexo', 'idade', 'raça', 'pais', 'curso']]
display(ano2019)
```

```python
# pegar 1 valor específico
print(df_lista.loc[1, 'pais'])
```

```python
# Adicionado 1 coluna
# a partir de uma coluna que existe 
df_lista['data_colacao_grau'] = df_lista['prazo_conclusao'] + 1
display(df_lista)
```

```python
# Criar uma coluna com valor padrão 
df_lista.loc[:, 'imposto'] = 0
display(df_lista)
```
