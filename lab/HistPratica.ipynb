### Pratica de Gráficos Histograma

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('lista_disc.csv')
dados.shape
```

```python
dados.head()
```

```python
plt.hist(dados.iloc[:,1], bins=6)
plt.title('Lista de Discentes')
plt.ylabel('Frequência')
plt.xlabel('Ano')
```

```python
agru = dados.groupby(['ano_ingresso']).size()
agru
```

```python
sns.distplot(dados.iloc[:,2], hist = True, kde = True, bins = 6, color = 'cyan',
            hist_kws={'edgecolor':'black'}
)
```

```python
plt.hist(dados.iloc[:,6], bins = 10)
plt.title('Lista de Discentes')
plt.ylabel('Frequência')
plt.xlabel('Idades')
```

```python
plt.hist(dados.iloc[:,8], bins = 10)
plt.title('Lista')
plt.ylabel('Frequência')
plt.xlabel('Pais')
```

```python
plt.hist(dados.iloc[:,9], bins = 20)
plt.title('Ávores')
plt.ylabel('Frequência')
plt.xlabel('Altura')
```
