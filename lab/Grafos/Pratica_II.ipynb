```python
# Importação das bibliotecas
from igraph import Graph
from igraph import plot
import numpy as np
import igraph
```

```python
grafo = Graph(edges=[(0,1),(1,2),(3,4),(0,2),(0,3)], directed=True)
grafo.vs['label'] = ['Joao', 'Pedro', 'Jose', 'Ana', 'Maria']
grafo.vs['peso'] = [50,40,20,25,35]
grafo.es['TipoAmizade'] = ['Amigo', 'Conhecido','MelhoAmigo','Inimigo','Colega']
grafo.es['weight'] = [1,2,1,3,4]
```

```python
print(grafo.vs[0])
```

```python
print(grafo.es[1])
```

```python
for v in grafo.vs:
    print(v)
```

```python
# Definição de cores para os vértices
grafo.vs['cor'] = ['cyan', 'blue', 'red', 'yellow', 'green']
plot(grafo, bbox=(0,0,600,600),vertex_color = grafo.vs['cor'])
```

```python
# pesos para as arrestas
plot(
    grafo, bbox=(600,600),  edge_width = grafo.es['weight'], 
    vertex_color = grafo.vs['cor']
)
```

```python
# Pesos para os vértices
plot(
    grafo, bbox=(600,600), vertex_size = grafo.vs['peso'],
    edge_width = grafo.es['weight'], vertex_color = grafo.vs['cor']
)
```
