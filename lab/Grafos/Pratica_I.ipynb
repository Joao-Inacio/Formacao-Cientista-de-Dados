```python
# Importação das bibliotecas
from igraph import Graph
from igraph import plot
```

```python
# Criando o grafo
grafo1 = Graph(edges = [(0,1),(1,2),(2,3),(2,15),(15,16),(16,3),(3,4),(4,5),
                        (5,6),(6,7),(7,8),(8,9),(9,10),(10,6),(10,4),(9,11),
                        (11,3),(11,12),(12,13),(13,14),(14,1)])
grafo1.vs['label'] = range(grafo1.vcount())
print(grafo1)
```

```python
# Gráfico
plot(grafo1, bbox = (0,0,900,900))
```
