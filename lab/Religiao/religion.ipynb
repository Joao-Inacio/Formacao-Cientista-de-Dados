```python
dataparse = lambda dates: datetime.strptime(dates, '%Y')
df = pd.read_csv(
    './data/global.csv', parse_dates = ['year'], index_col = 'year',
    date_parser= dataparse
)
df
```

```python
ts = df['christianity_romancatholic']
ts
```

```python
# Visualização da série temporal completa
plt.plot(ts)
```

```python
# Visualização por ano
ts_ano = ts.resample('A').sum()
plt.plot(ts_ano)
ts_ano
```
