### Formação Cientista de Dados
* Histograma com Densidade e Seabord

```python
# Importação das Libs
import pandas as pd 
import seaborn as srn
import matplotlib.pyplot as plt 
```

```python
# Base de dados
base = pd.read_csv('./data/trees.csv')
base.head()
```

```python
# Histograma com 10 divisãoes (bins) e com gráfico de densidade
srn.histplot(
    base.Volume, bins=10, kde=True, stat='density', color='cyan'
).set_title('Volume');
```
