#### Formação Cientista de Dados
* Divisão de tela (subgráficos)

```python
# Importação das bibliotecas 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Base de dados
base = pd.read_csv('./data/trees.csv')
base.head()
```

```python
# Girth com volume
plt.scatter(base.Girth, base.Volume);
```

```python
# Girth com heigth
plt.scatter(base.Girth, base.Height);
```

```python
# Heigth com Volume
plt.scatter(base.Height, base.Volume, marker='*');
```

```python
# Histograma volume
plt.hist(base.Volume);
```

```python
# imprimindo juntos
# Criação de figuras, no qual os gráficos serão posicionados
plt.figure(1)
plt.subplot(2,2,1) # (Linha, coluna, posição)
plt.scatter(base.Girth, base.Volume)
plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height)
plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume, marker='*')
plt.subplot(2,2,4)
plt.hist(base.Volume);
```
