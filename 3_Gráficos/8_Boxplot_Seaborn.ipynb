#### Formação Cientista de Dados
* Boxplot com Seaborn

```python
# Importação das libs
import pandas as pd 
import seaborn as srn
```

```python
# Base de Dados
base = pd.read_csv('./data/trees.csv')
base.head()
```

```python
# Visualização de um boxplot
srn.boxplot(data=base.Volume, orient='h').set_title('Árvores');
```

```python
# Visualização de vários boxplot na mesma imagem
srn.boxplot(data=base);
```
