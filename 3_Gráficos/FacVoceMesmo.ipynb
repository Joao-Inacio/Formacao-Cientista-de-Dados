```python
# Importando as biblioteca
import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as srn
import numpy as np
```

```python
# base de dados
dados = pd.read_csv('dados.csv', sep=';')
dados.head()
```

```python
v = np.histogram(dados.iloc[:,3], bins = 6)
v
```

```python
p = np.histogram(dados.iloc[:,2], bins = 6)
p
```

```python
# Visualização do histograma com 6 divisões (bins)
plt.hist(dados.iloc[:,2], bins = 6)
plt.title('Dados')
plt.ylabel('Frequência')
plt.xlabel('PIB')
```

```python
agru = dados.groupby(['VALOREMPENHO'])['PIB']
agru
```

```python
srn.distplot(dados.PIB, bins=10, axlabel = 'PIB').set_title('PIB')
```

```python
srn.distplot(dados.VALOREMPENHO, bins=10, axlabel = 'VALOR do EMPENHO').set_title('EMPENHO')
```

```python
agru = dados.sort_values('PIB').head(10)
agru = agru.iloc[:,1:3]
agru
agru.plot.bar(x='MUNICIPIO',y='PIB',color='cyan')
```

```python
agru = dados.sort_values('VALOREMPENHO').head(10)
agru = agru.iloc[:,[1,3]]
agru
agru.plot.bar(x='MUNICIPIO',y='VALOREMPENHO',color='cyan')
```
