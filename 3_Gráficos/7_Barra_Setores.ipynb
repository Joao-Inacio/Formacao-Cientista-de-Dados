#### Formação Cientista de Dados
* Gráfico de barras e setores

```python
# Importação de libs
import pandas as pd 
```

```python
# Base de Dados
base = pd.read_csv('insect.csv')
base.shape
```

```python
# Dados
base.head()
```

```python
# Agrupamento dos dados baseado do atributo 'spray', contando e somando os registros
agru = base.groupby(['spray'])['count'].sum()
agru
```

```python
# Gráfico de Barra
agru.plot.bar(color='gray')
```

```python
# Cores
agru.plot.bar(color = ['blue','yellow','red','green','pink','orange',])
```

```python
# Gráfico de pizza
agru.plot.pie()
```

```python
# Com Legenda
agru.plot.pie(legend = True)
```
